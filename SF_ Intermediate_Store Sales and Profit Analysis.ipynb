```python
print(df.isnull().sum())
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
print(df.dtypes)
```

```python
sales=df['Sales'].sum()
profit = df['Profit'].sum()
print(f"Total Sales: Rupees {sales}, Total Profit: Rupees {profit}")
```

```python
category=df.groupby('Category')['Sales'].sum().sort_values()
category.plot(kind='bar', title='Sales by Category', color='yellow')
```

```python
Rprofit = df.groupby('Region')['Profit'].sum().sort_values()
Rprofit.plot(kind='bar', title='Profit by Region', color='pink')
```

```python
top = df.groupby('Product Name')['Sales'].sum().sort_values().head(10)
top.plot(kind='bar', title='Top Selling Products', color='lavender')
```
